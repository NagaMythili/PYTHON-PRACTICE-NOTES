# DATA FRAMES

A DATA FRAME is a 2 dimensional data structure in which we store the data in form of "TABLES" i.e in the view of [rows X columns]           
ex: (4,5) # 4-->row indication, 5-->column indication

# FEATURES OF DATA FRAMES :

      -->potentially columns of different data types.
      -->Size is Mutable
      -->Its Contains only Labelled Data Axeses(Row,Column)
      -->we can perform Arithmetic Operations
        
    "we can Create a Table via DATA FRAME that is known as DATA SET"

# WHAT IS DATA SET ?

 A DataSet is another name of table,in which we store the data in the form of Rows & Columns.The data is Mutable,Updatable.The data set Size is Negotiable.

STRUCTURE OF DATAFRAME/DATASET :
    
    1.ROWS
    2.COLUMNS
    3.CELLS

# STRUCTURE OF DATAFRAME/DATASET :
    
    1.ROWS --->(R1,R2,R3)
    2.COLUMNS--->(C1,C2,C3)
    3.CELLS---->ONE BLOCK BY INDICATION WHICH ROW AND COLUMN AS SHOWN BELOW PICTURE.

# SYNTAX OF DATA FRAME:

pandas.DataFrame(data,index,columns,dtype,copy)
DATA-->Info , INDEX-->Val ,COLUMNS--> Operations on it indicating , dtype--> Data type specifing , copy-->dublication

# CREATING THE DATAFRAME USING "LIST"

```python
import pandas as n
a=[10,20,30,40,50]
b=n.DataFrame(a)
print(b)
```

# CREATING THE DATAFRAME USING "DICTIONARY"

```python
import pandas as n
a=(10,20,30,40,50)
b=n.DataFrame(a)
print(b)   # due to 2d this will return in rows and columns .
```

# CREATING THE DATAFRAME USING "SERIES"

```python
import pandas as n
a=[10,20,30,40,50,60]
b=n.Series(a)
c=n.DataFrame(b)
print(c)
```

# CREATING DATAFRAME BY USING "GENERAL WAY"

```python
import pandas as pd
ds={'S.NO':pd.Series([1,2,3,4,5]),
    'ROLL NO':pd.Series([1001,1002,1003,1004,1005]),
    'NAME':pd.Series(['MYTHS','RAM','SITA','RUSHI','DEV']),
    'SUB1':pd.Series([50,30,40,50,50]),
    'SUB2':pd.Series([50,20,70,80,50])}
df=pd.DataFrame(ds)
print(df)
```
