# COVARIANCE   ---> cov()

---> THE RELATION BETWEEN TWO DATA MEMBERS WHICH ARE IN TWO DIFFERENT ATTRIBUTES. 

---> IT IS A MEASUREMENT OF CORRELATION.       

---> INDIVIDUAL ELEMENTS OF THE DATA FROM 2 COLUMNS. 

---> THE RESULT MAY VARY EVERY SECOND.       

---> IT'S RANGE IN BETWEEN -∞ TO +∞              

```python
import pandas as pd
ds={'S.NO':pd.Series([1,2,3,4,5]),
    'ROLL NO':pd.Series([1001,1002,1003,1004,1005]),
    'NAME':pd.Series(['MYTHS','RAM','SITA','RUSHI','DEV']),
    'SUB1':pd.Series([50,30,40,50,50]),
    'SUB2':pd.Series([50,20,70,80,50]),
   'Grades':pd.Series(['B','C','A','A+','B'])}
df=pd.DataFrame(ds)
df['RESULT']=df['SUB1']+df['SUB2']
print(df)
print(df['SUB1'].cov(df['SUB2']))
print(df['SUB2'].cov(df['SUB1']))  # NO CHANGE IN THE VALUES BECAUSE HERE IT TAKES AVERAGE OF THE VALUES.
```
