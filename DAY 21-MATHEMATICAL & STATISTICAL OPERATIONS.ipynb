# PANDAS COMPUTATIONAL TOOLS FOR STATISTICAL COMPUTATION OR MATHEMATICAL OPERATIONS
--A computation is a process or completing a task by using many operations mathematically or statistically.

# we have top 5 computational tools for Statistics    
1.min()   
2.max()   
3.rank()   
4.corr()  --co-relation  
5.cov()   --co-varience

```python
import pandas as pd
ds={'S.NO':pd.Series([1,2,3,4,5]),
    'ROLL NO':pd.Series([1001,1002,1003,1004,1005]),
    'NAME':pd.Series(['MYTHS','RAM','SITA','RUSHI','DEV']),
    'SUB1':pd.Series([50,30,40,50,50]),
    'SUB2':pd.Series([50,20,70,80,50]),
   'Grades':pd.Series(['B','C','A','A+','B'])}
df=pd.DataFrame(ds)
#print(df)
'''print("----------------------------------------------")
print("----ROW ADDITION-----")
df.loc[5]=[6,1006,'SRIKALA',60,80,'A+']'''
df['RESULT']=df['SUB1']+df['SUB2']
print(df)
```

# min()
---which returns minimum value in a column.

```python
mini=df['RESULT'].min()
print("Minimum Value  : ",mini)
```

# max()   
 ---which returns maximum value in a column.

```python
maxi=df['RESULT'].max()
print("Maximum Value  : ",maxi)
```

# rank()  
--By using this method we can rank all the data Members in a column in a Ascending order if those are in Numerics,Come to Alphabets.It follows Alphabetic Order.

```python
print(df['RESULT'].rank())
```

```python
df.rank()
```

```python
print(df)
print("---------------------------------------------------------")
print("----PRINTING THE VALUES AS PER THE RANK IN ORDERWISE-----")
print("\n")
print(df.rank())
```

# Correlation ----> corr()    
It is a Relation between two Column Data Members in a Dataset.     
2.We use a Method called "corr()"        
3.It is a Scaled Form of a Covariance.      
4.The Correlation Values are lies between " -1 and +1 ".         
                                            ----------    
          SYNTAX: attribute1.corr(attribute2)

# TYPES OF CORRELATIONS:       
    1.POSITIVE(0 TO 1 in between values we got means positive)   
    
           ---if values in the range of 0 to +0.5 then it said to be "High positive".  
       
           ---if values in the range of +0.5 to +1 then it is said to be "Low positive".
      
    2.NEGATIVE(-1 to 0 in between we got means negative)    
    
           ---if values in the range of 0 to -0.5 then it is said to be "High negative".  
         
           ---if values in the range of -0.5 to -1 then it is said to be "Low negative".   
        
    3.NO CORRELATION

```python
x=df['SUB1'].corr(df['SUB2'])
print(x)
```

```python
import pandas as pd
ds={'S.NO':pd.Series([1,2,3,4,5]),
    'ROLL NO':pd.Series([1001,1002,1003,1004,1005]),
    'NAME':pd.Series(['MYTHS','RAM','SITA','RUSHI','DEV']),
    'SUB1':pd.Series([50,30,40,50,50]),
    'SUB2':pd.Series([50,20,70,80,50]),
   'Grades':pd.Series(['B','C','A','A+','B'])}
df=pd.DataFrame(ds)
#print(df)
'''print("----------------------------------------------")
print("----ROW ADDITION-----")
df.loc[5]=[6,1006,'SRIKALA',60,80,'A+']'''
df['RESULT']=df['SUB1']+df['SUB2']
print(df)
```

```python
x=df['SUB1'].corr(df['RESULT'])  # USING DATASET CORRELATION BETWEEN 2 ATTRIBUTES.
print(x)
```
