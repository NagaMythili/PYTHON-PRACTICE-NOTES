# Clustering  
   ---> grouping of the data into thier own Property.

```python
import pandas as pd
ds={'S.NO':pd.Series([1,2,3,4,5]),
    'ROLL NO':pd.Series([1001,1002,1003,1004,1005]),
    'NAME':pd.Series(['MYTHS','RAM','SITA','RUSHI','DEV']),
    'SUB1':pd.Series([50,30,40,50,50]),
    'SUB2':pd.Series([50,20,70,80,50]),
   'Grades':pd.Series(['B','C','A','A+','B'])}
df=pd.DataFrame(ds)
df['RESULT']=df['SUB1']+df['SUB2']
print(df)
```

# Summarization  
   ---> Details about Dataset    
    here we use 2 methods                
    1. info()            
    2.describe()                       

```python
import pandas as pd
ds={'S.NO':pd.Series([1,2,3,4,5]),
    'ROLL NO':pd.Series([1001,1002,1003,1004,1005]),
    'NAME':pd.Series(['MYTHS','RAM','SITA','RUSHI','DEV']),
    'SUB1':pd.Series([50,30,40,50,50]),
    'SUB2':pd.Series([50,20,70,80,50]),
   'Grades':pd.Series(['B','C','A','A+','B'])}
df=pd.DataFrame(ds)
df['RESULT']=df['SUB1']+df['SUB2']
print(df)
print("\n")
print("------------info------------------------")
print(df.info())
print("\n")
print("------------describe------------------------------------")
print(df.describe())
```
