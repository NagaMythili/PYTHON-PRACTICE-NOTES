# CLASSIFICATION  
   ---> It is a predictive model and focus on values present in the Dataset.

```python
import pandas as pd
ds={'S.NO':pd.Series([1,2,3,4,5]),
    'ROLL NO':pd.Series([1001,1002,1003,1004,1005]),
    'NAME':pd.Series(['MYTHS','RAM','SITA','RUSHI','DEV']),
    'SUB1':pd.Series([50,30,40,50,50]),
    'SUB2':pd.Series([50,20,70,80,50]),
   'Grades':pd.Series(['B','C','A','A+','B'])}
df=pd.DataFrame(ds)
df['RESULT']=df['SUB1']+df['SUB2']
print(df)
```

For same datatype values only we perform the classification as we see above dataset if we select one cloumn which is the RESULT. so for classification we take some constraints or conditions to classify that similar datatype values.
coming to here if we consider this RESULT Column:
    
    constraints are taken as:
    
    1. which is >100 -->(are 110,130  from the result column)
    2. which is =100 -->(are 100,100 from the result column)
    3. which is <100 -->(are 50 from the result column)
    
by taking any reference value to classify the data from the dataset.
