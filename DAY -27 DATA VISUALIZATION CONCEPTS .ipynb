# 1. Normal Distribution

```python
import numpy as np
x=np.random.normal(1,2,10) # range is from 1 to 2 in between and 10 values
print(x)
```

```python

import numpy as np
x=np.random.normal(10,20,10) # range is from 10 to 20 in between and 10 values
print(x)
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.normal(10,20,10)    
print(x)
plt.hist(x,100)
plt.show()
```

 it acts as margins if data values range >20 and <10

# 2. UNIFORM DISTRIBUTION

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.uniform(10,20,10)    
print(x)
plt.hist(x,100)
plt.show()
```

# 3. LOGISTIC DISTRIBUTION

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.logistic(10,20,10)    
print(x)
plt.hist(x,100)
plt.show()
```
